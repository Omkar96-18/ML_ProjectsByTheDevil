```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('../../../datasets/tmdb_5000_movies.csv')
credits = pd.read_csv('../../../datasets/tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on='title')
movies.info()
```

```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords','cast','crew', 'popularity']]
```

```python
movies.sample(10)
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies.dropna(inplace=True)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.sample(4)
```

```python
def convertCast(text):
    L = []
    cnt = 0
    for i in ast.literal_eval(text):
        if cnt < 5:
            L.append(i['name'])
        cnt+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convertCast)
```

```python
def fetchDirector(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetchDirector)
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['genres']
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.sample(3)
```

```python
movies.iloc[0]
```

```python
movies['cast'].apply(lambda x:x[0:5])
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ", ""))
    return L1
```

```python

movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies.drop(columns=['overview','genres','keywords','cast','crew','popularity'])
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head(5)
```

```python
new_df['tags'][1]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head(5)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector[0]
```

```python
list(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
new_df[new_df['title'] == 'Batman Begins'].index[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])
```

```python
def recommend(movie):
    idx = new_df[new_df['title'] == movie].index[0]
    distance = sorted(list(enumerate(similarity[idx])), reverse=True, key=lambda x:x[1])
    for i in distance[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```
